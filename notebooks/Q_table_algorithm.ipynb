```python
def plot_rewards(rewards_given, episode, window_size=10):
    episodes = range(episode + 1)
    rewards_moving_avg = np.convolve(rewards_given[:episode + 1], np.ones(window_size)/window_size, mode='valid')
    plt.plot(episodes[window_size-1:], rewards_moving_avg)
    plt.xlabel('Episode')
    plt.ylabel('Moving Average Rewards')
    plt.title(f'Moving Average Rewards per Episode (Window Size = {window_size})')
    plt.grid(True)
    plt.show()
```

```python
def simulate_key_press(key):
    pygame.event.post(pygame.event.Event(KEYDOWN, {'key': key}))
```

```python
def perform_action(action):
    if action==0:
        simulate_key_press(K_UP)
    if action==1:
        simulate_key_press(K_RIGHT)
    if action==2:
        simulate_key_press(K_DOWN)
    if action==3:
        simulate_key_press(K_LEFT)
    if action==5:
        simulate_key_press(K_SPACE)
```

```python
# Training parameters
learning_rate = 0.001
discount_factor = 0.99

epsilon = 1
epsilon_decay=0.997
epsilon_min=0.01

total_episodes = 10000

num_distances_manhattan=41
action_space_size=6

visualisation_frequency=400
```

```python
agent= QTable(num_distances_manhattan,action_space_size,num_orientations_kronecker=2)
rewards_given=np.zeros(total_episodes)
```

```python
# Initialize pygame and the environment
pygame.init()


for episode in range (total_episodes):
    env = TankEnv()
    env.reset()

    # Set up the display
    screen_size = (env.max_x * 20, env.max_y * 20)  # Scale up the game screen
    screen = pygame.display.set_mode(screen_size)
    clock = pygame.time.Clock()

    current_state=env.state
    current_distance,current_orientation=grab_distance_and_kronecker(current_state)
    done=False
    total_reward = 0

    while not done:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                done = True
        action=agent.choose_action(current_distance, current_orientation, epsilon)

        #perform action
        perform_action(action) 

        # Update the environment
        next_state, reward, done, _= env.step(action)
        
        # Render the game state
        frame = env.render()
        frame = np.repeat(np.repeat(frame, 18, axis=0), 18, axis=1)  # Scale up the frame for visibility
        surface = pygame.surfarray.make_surface(frame)
        screen.blit(surface, (0, 0))
        pygame.display.flip()

        # Cap the frame rate
        clock.tick(9)
        
        next_distance,next_orientation=grab_distance_and_kronecker(next_state)

        #update q_table
        agent.update_q_value( current_distance, current_orientation, action, learning_rate, reward, discount_factor, next_distance, next_orientation)
        total_reward += reward

        current_state=next_state
        current_distance=next_distance
        current_orientation=next_orientation

    rewards_given[episode]=total_reward
    
    epsilon=max(epsilon_min,epsilon*epsilon_decay)

    if episode%visualisation_frequency==0 and episode>10:
        plot_rewards(rewards_given,episode,window_size=10)
    
pygame.quit()

```
